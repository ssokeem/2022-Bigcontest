```python
import pandas as pd
import numpy as np
```

```python
cafe = pd.read_csv('./cluster/cafe_cluster_3.csv', encoding="UTF-8")
food = pd.read_csv('./cluster/food_cluster_3.csv', encoding="UTF-8")
```

## 카페

```python
cafe
```

```python
cafe_top20 = cafe.sort_values('인기지수', ascending=False).head(20)
```

```python
cafe_top20
```

```python
cafe_eco = pd.read_csv('./etc_data/cafe_eco.csv', encoding="UTF-8")
cafe_eco['place'] = cafe_eco['place'].str.replace(' ', '')
```

```python
cafe_top20_key = pd.merge(cafe_top20, cafe_eco, left_on='place', right_on='place', how='left')
```

```python
cafe_top20_key
```
