```python
import pandas as pd
import numpy as np
import re, string
```

# 크롤링 데이터

```python
cafe = pd.read_csv('./naver_review/cafe.csv', encoding='UTF-8')
foods = pd.read_csv('./naver_review/foods.csv', encoding='UTF-8')
motel = pd.read_csv('./naver_review/motel.csv', encoding='UTF-8')
pension = pd.read_csv('./naver_review/pension.csv', encoding='UTF-8')
resort = pd.read_csv('./naver_review/resort.csv', encoding='UTF-8')
tour = pd.read_csv('./naver_review/tour.csv', encoding='UTF-8')
hotel = pd.read_csv('./naver_review/hotel.csv', encoding='UTF-8')
```

## 카페

```python
cafe.head()
```

```python
def review_num(text):
    if text.find('방문자리뷰') == -1:
        real_num = 0
    else:
        start_idx = text.find('방문자리뷰') + 6
        last_idx = text.find('\'', start_idx)
        review_num = text[start_idx:last_idx]
        real_num = re.sub(r"[^0-9]", "", review_num)
        if real_num.strip() == "":
            real_num = 0
    return real_num

def preprocessing(data):
    data['Naver_review_num'] = data['Naver_review'].apply(lambda x: review_num(x))
    new_data = data.drop(['Unnamed: 0', '별점', '별점갯수', 'Naver_review'], axis=1)
    return new_data
```

```python
cafe_review = preprocessing(cafe)
```

```python
cafe_review = cafe_review.rename(columns={'카페명':'place'})
cafe_review
```

## 식당

```python
food_review = preprocessing(foods)
```

```python
food_review = food_review.rename(columns={'카페명':'place'})
food_review
```

## 모텔

```python
motel_review = preprocessing(motel)
motel_review = motel_review.rename(columns={'카페명':'place'})
motel_review
```

## 펜션

```python
pension_review = preprocessing(pension)
pension_review = pension_review.rename(columns={'카페명':'place'})
pension_review
```

## 리조트

```python
resort_review = preprocessing(resort)
resort_review = resort_review.rename(columns={'카페명':'place'})
resort_review
```

## 관광지

```python
tour.head()
```

```python
tour_review = tour.drop(['Unnamed: 0', '별점', '별점갯수'], axis=1)
tour_review = tour_review.rename(columns={'카페명':'place', 'Naver_review':'Naver_review_num'})
tour_review
```

```python
tour_review[tour_review['place'].str.contains('시장', na = False)]
```

## 호텔

```python
hotel_review = preprocessing(hotel)
hotel_review = hotel_review.rename(columns={'카페명':'place'})
hotel_review
```

## CSV 파일로 내보내기

```python
cafe_review.to_csv('./naver_review/cafe_review.csv', header=True, index=False, encoding="UTF-8")
food_review.to_csv('./naver_review/food_review.csv', header=True, index=False, encoding="UTF-8")
motel_review.to_csv('./naver_review/motel_review.csv', header=True, index=False, encoding="UTF-8")
pension_review.to_csv('./naver_review/pension_review.csv', header=True, index=False, encoding="UTF-8")
resort_review.to_csv('./naver_review/resort_review.csv', header=True, index=False, encoding="UTF-8")
tour_review.to_csv('./naver_review/tour_review.csv', header=True, index=False, encoding="UTF-8")
hotel_review.to_csv('./naver_review/hotel_review.csv', header=True, index=False, encoding="UTF-8")
```
