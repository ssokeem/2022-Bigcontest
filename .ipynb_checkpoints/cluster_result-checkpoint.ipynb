```python
import pandas as pd
import numpy as np
```

```python
place_location = pd.read_csv('./place_address/place_location.csv', encoding="UTF-8")
cluster_base = pd.read_csv('./place_address/cluster.csv', encoding="UTF-8")
```

```python
place_location['k_means_cluster'] = cluster_base['k_means_cluster']
```

```python
place_location
```

```python
categories = ['cafe', 'motel', 'hotel', 'pension', 'resort', 'food', 'tour']

for category in categories:
    globals()['{}_popularity'.format(category)] = pd.read_csv('./review_popularity/' + '{}_popularity'.format(category)+ '.csv', encoding='UTF-8')
```

```python
place_cafe = place_location[place_location['category']=='카페']
place_motel = place_location[place_location['category']=='모텔']
place_hotel = place_location[place_location['category']=='호텔']
place_pension = place_location[place_location['category']=='펜션']
place_resort = place_location[place_location['category']=='리조트']
place_food = place_location[place_location['category']=='식당']
place_tour = place_location[place_location['category']=='관광지']
```

```python
def matching(place_data, poupularity_data):
    place_data['인기지수'] = np.nan
    
    place_data = place_data.reset_index()
    place_data = place_data.drop('index', axis=1)
    
    place_data['place'] = place_data['place'].str.replace(' ', '')
    
    for i in range(len(place_data)):
        place = place_data['place'][i]
        pop = poupularity_data[poupularity_data['place']==place]['인기지수'].astype(float)
        place_data['인기지수'][i] = pop
        
    return place_data
```

```python
cafe_cluster = matching(place_cafe, cafe_popularity)
motel_cluster = matching(place_motel, motel_popularity)
hotel_cluster = matching(place_hotel, hotel_popularity)
resort_cluster = matching(place_resort, resort_popularity)
pension_cluster = matching(place_pension, pension_popularity)
tour_cluster = matching(place_tour, tour_popularity)
food_cluster = matching(place_food, food_popularity)
```

## 클러스터별 분할

```python
def set_cluster(data):
    c0 = data[data['k_means_cluster']==0]
    c1 = data[data['k_means_cluster']==1]
    c2 = data[data['k_means_cluster']==2]
    c3 = data[data['k_means_cluster']==3]
    return c0, c1, c2, c3
```

```python
cafe_cluster_0, cafe_cluster_1, cafe_cluster_2, cafe_cluster_3 = set_cluster(cafe_cluster)
food_cluster_0, food_cluster_1, food_cluster_2 ,food_cluster_3 = set_cluster(food_cluster)
tour_cluster_0, tour_cluster_1, tour_cluster_2, tour_cluster_3 = set_cluster(tour_cluster)
hotel_cluster_0, hotel_cluster_1, hotel_cluster_2, hotel_cluster_3 = set_cluster(hotel_cluster)
motel_cluster_0, motel_cluster_1, motel_cluster_2, motel_cluster_3 = set_cluster(motel_cluster)
resort_cluster_0, resort_cluster_1, resort_cluster_2, resort_cluster_3 = set_cluster(resort_cluster)
pension_cluster_0, pension_cluster_1, pension_cluster_2, pension_cluster_3 = set_cluster(pension_cluster)
```

```python
for category in categories:
    for i in range(0, 4):
        globals()['{}_cluster_{}'.format(category, i)].to_csv('./cluster/' + '{}_cluster_{}'.format(category, i) + '.csv', header=True, index=False, encoding="UTF-8")
```

## 친환경 매장 찾기

```python
place_location['place'] = place_location['place'].str.replace(' ', '')
place_location['property'] = 0
```

```python
zero_waste = ['알맞은시간', 'AND유CAFE', '제주소녀', '발트글라스', '고집돌우럭', '펜고호다', '어나더페이지', '북스페이스곰곰', '지구별가게', '꽃마리협동조합', '핸드메이드라이프', '제주용기']
green_store = ['롯데마트','뉴월드마트', '마트로', '바이올가', '제주용기']
```

```python
for i in range(len(place_location)):
    name = place_location['place'][i]
    for store in zero_waste:
        if name.find(store) > -1 or name == store:
            place_location['property'][i] = 1
    for store in green_store:
        if name.find(store) > -1 or name == store:
            place_location['property'][i] = 1
```
