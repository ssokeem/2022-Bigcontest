### 라벨링

```python
tour_keyword = pd.read_csv('./kakao_review/tour_keyword.csv', encoding="UTF-8")
tour_keyword['place'] = tour_keyword['place'].str.replace(' ', '')
```

```python
tour_sort = tour.sort_values('인기지수', ascending=False)
tour_sort = tour_sort.reset_index()
tour_sort.drop('index', axis=1, inplace=True)
```

```python
tour_top20 = tour.sort_values('인기지수', ascending=False).head(20)
tour_keyword = tour_keyword[['place', '환경', '친환경']]
tour_key = pd.merge(tour_sort, tour_keyword, left_on='place', right_on='place', how='left')
```

```python
tour_key_100 = tour_key[:100]
tour_key_100
```
