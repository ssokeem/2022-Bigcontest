```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler
```

```python
categories = ['cafe', 'motel', 'hotel', 'pension', 'resort', 'food', 'tour']
```

```python
for category in categories:
    globals()['{}'.format(category)] = pd.read_csv('./final_review/' + '{}_result'.format(category)+ '.csv', encoding='UTF-8')

tour['insta'] = tour['insta'].apply(lambda x: str(x))
```

```python
def popular_indicator(data):
    data['insta'] = data['insta'].str.replace(',','').astype(int)
    data = data.replace(0, np.NaN)
    data['insta'] = data['insta'].fillna(0)
    
    data['mean_rate'] = data[['Naver_rate', 'kakao_rate']].mean(axis=1)
    data2 = data.drop(['place', 'Naver_rate', 'kakao_rate'], axis=1)
    
    scaler = MinMaxScaler()
    df = scaler.fit_transform(data2)
    df = pd.DataFrame(df, columns=data2.columns)
    
    df['mean_num'] = df[['Naver_review_num', 'kakao_review_num']].mean(axis=1)
    df.drop(['Naver_review_num', 'kakao_review_num'], axis=1, inplace=True)
    
    df.insert(0, 'place', data['place'])
    df['인기지수'] = df[['insta', 'mean_rate', 'mean_num']].mean(axis=1)
    
    return df
```

```python
cafe_popularity = popular_indicator(cafe)
motel_popularity = popular_indicator(motel)
hotel_popularity = popular_indicator(hotel)
pension_popularity = popular_indicator(pension)
resort_popularity = popular_indicator(resort)
food_popularity = popular_indicator(food)
tour_popularity = popular_indicator(tour)
```

```python
for category in categories:
    globals()['{}_popularity'.format(category)].to_csv('./review_popularity/' + '{}_popularity'.format(category) + '.csv', header=True, index=False, encoding="UTF-8")
```
