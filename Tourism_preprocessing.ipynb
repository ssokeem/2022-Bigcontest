## 인당 평균 입장 가격

```python
enter_NUM = enter.drop(['CL_NM', 'ADDR', 'TEL_NO', 'SETLE_PRICE', 'ENTRN_DE'], axis=1)
```

```python
enter_group = enter.groupby(['TRRSRT_NM', 'ADDR']).sum()
```

```python
enter_group['price_per_person'] = round(enter_group['SETLE_PRICE']/enter_group['TOT_ENTRN_NMPR_CO'],1)
```

```python
enter_group
```

```python
enter_TOT = enter_group.drop(['SETLE_PRICE', 'price_per_person'], axis=1)
enter_TOT = enter_TOT.T.unstack().reset_index()
enter_TOT.drop('level_3', axis=1, inplace=True)
enter_TOT.rename(columns = {0:'TOT_ENTRN_NMPR_CO'}, inplace=True)
```

```python
enter_PRICE = enter_group.drop(['TOT_ENTRN_NMPR_CO', 'price_per_person'], axis=1)
enter_PRICE = enter_PRICE.T.unstack().reset_index()
enter_PRICE.drop('level_3', axis=1, inplace=True)
enter_PRICE.rename(columns = {0:'SETLE_PRICE'}, inplace=True)
```

```python
enter_PERSON = enter_group.drop(['TOT_ENTRN_NMPR_CO', 'SETLE_PRICE'], axis=1)
enter_PERSON = enter_PERSON.T.unstack().reset_index()
enter_PERSON.drop('level_3', axis=1, inplace=True)
enter_PERSON.rename(columns = {0:'PRICE_PER_PERSON'}, inplace=True)
```

```python
v1 = pd.merge(enter_TOT, enter_PRICE, how='inner', on=['CL_NM', 'TRRSRT_NM', 'ADDR'])
```

```python
enter_final = pd.merge(v1, enter_PERSON, how='inner', on=['CL_NM', 'TRRSRT_NM', 'ADDR'])
```

```python
enter_final.to_csv('./etc_data/enter_final.csv', header=True, index=False, encoding="cp949")
```
