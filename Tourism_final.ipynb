```python
unique_nm = carbon['TRRSRT_NM'].unique()
```

```python
len(unique_nm)
```

```python
count = [0] * len(unique_nm)

for i in range(len(unique_nm)):
    for j in range(len(carbon)):
        if unique_nm[i] == carbon['TRRSRT_NM'][j]:
            count[i] += 1
        else:
            pass
```

```python
idx = []

for i in range(len(count)):
    if count[i] > 1:
        idx.append(i)
```

```python
rep_nm = []

for i in idx:
    rep_nm.append(unique_nm[i])
```

```python
rep_nm
```

```python
carbon['Duplicate_status'] = False
```

```python
for i in range(len(carbon)):
    if carbon['TRRSRT_NM'][i] in rep_nm:
        carbon['Duplicate_status'][i] = True
```

```python
carbon_rep = carbon[carbon['Duplicate_status'] == True]
```

```python
carbon_rep = carbon_rep.sort_values('TRRSRT_NM')
```

```python
carbon_rep
```

```python
avg['Duplicate_status'] = False

for i in range(len(avg)):
    if avg['TRRSRT_NM'][i] in rep_nm:
        avg['Duplicate_status'][i] = True
```

```python
avg_rep_df = avg[avg['Duplicate_status']==True]
```

```python
carbon
```

```python
avg_rep_df['AVERAGE_VISITOR'] = round(avg_rep_df['AVERAGE_VISITOR'], 1)
```

```python
avg_rep_df
```

```python
for i in range(len(carbon)):
    if carbon['Duplicate_status'][i] == True:
        nm = carbon['TRRSRT_NM'][i]
        avg_visit = avg_rep_df[avg_rep_df['TRRSRT_NM'] == nm]['AVERAGE_VISITOR']
        carbon['AVERAGE_VISITOR'] = float(avg_visit)
```
